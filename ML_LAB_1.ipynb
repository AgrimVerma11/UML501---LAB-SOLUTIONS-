## Note - 

### Ravel -

- numpy.ravel() returns a contiguous flattened array.
- It converts an input array of any shape into a 1-dimensional array.
- It tries to return a view of the original array whenever possible to avoid copying data (which means it is memory-efficient).
- If a view is not possible (due to how the array is stored in memory), it returns a copy.

#### Key points - 

- Returns a 1-D array: Regardless of the input array’s shape, the output is always 1-dimensional.
- View vs Copy: ravel() attempts to return a view. If the input array’s memory layout allows it, the returned array shares the same data buffer. Changes to the returned array affect the original array.
- Difference with flatten(): .flatten() always returns a copy, whereas ravel() returns a view when possible.
- Order parameter matters: Changing the order affects how elements are read from the multi-dimensional array and thus changes the sequence of elements in the flattened array.
- Useful for reshaping without copying data unnecessarily, especially in large datasets where memory usage is critical.

```python
import cv2
import numpy as np

def img_to_array(path):
    # Read the image using OpenCV
    img = cv2.imread(path)

    # Handle invalid path or unsupported file
    if img is None:
        print("Image not found or unsupported format.")
        return None

    # Grayscale Image Case
    if len(img.shape) == 2:
        print("Detected Grayscale image.")
        np.savetxt("grayscale_image.txt", img, fmt='%d')
        return img.shape  # Return shape to be used later

    # RGB Image Case
    elif len(img.shape) == 3 and img.shape[2] == 3:
        print("Detected RGB image.")
        # Convert BGR (OpenCV default) to RGB
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Save reshaped RGB array to a .txt file
        reshaped = img_rgb.reshape(-1, 3)  # (H×W, 3)
        np.savetxt("rgb_image.txt", reshaped, fmt='%d')

        return img_rgb.shape  # Return (height, width, 3)

    else:
        print("Unsupported image format.")
        return None
```

```python
shape = img_to_array(r"E:\AI_ML\SEM 5\Lab Sessions\Agrim.jpg")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load RGB data from saved text file
data = np.loadtxt("rgb_image.txt", dtype=int)

# Unpack the original shape returned earlier
height, width, channels = shape  # channels = 3

# Reshape the flat RGB data back to image format
img_loaded = data.reshape((height, width, channels))

# Display the image
plt.imshow(img_loaded)
plt.axis('off')
plt.title('Reconstructed RGB Image from Text File')
plt.show()
```

```python
# Load grayscale image array
gray_img = np.loadtxt("rgb_image.txt", dtype=int)

# Display the grayscale image
plt.imshow(gray_img, cmap='gray')
plt.axis('off')
plt.title('Reconstructed Grayscale Image from Text File')
plt.show()
```
